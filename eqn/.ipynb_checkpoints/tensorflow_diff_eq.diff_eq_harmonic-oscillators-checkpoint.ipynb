```python
import tensorflow_diff_eq.diff_eq as diff_eq
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
eq = diff_eq.DifferentialEquation()

# Define the quantities involved in the harmonic oscillator
(x, k, m) = eq.generate_quantities(3)

# Define these quantities
# eq.define_quantity( quantity_name, initial_value, order_of_derivative, definition_derivative ) ( #the_equation_tself )
eq.define_quantity(k, 3.0, 0, 1)(0.0)
eq.define_quantity(m, 0.7, 0, 1)(0.0)

# The actual equation
eq.define_quantity(x, 1.0, 0, 2)(-k.d(0)*x.d(0)/m.d(0))

# Start tensorflow
sess = tf.Session()
sess.run(tf.global_variables_initializer())
simulate_op = eq.generate_simulate_operation(0.001)


# Simulate and record values
xs = np.zeros(5000)
ts = np.zeros(5000)
for i in range(5000):
    xs[i] = sess.run(x.d(0))
    if i < 4999:
        sess.run(simulate_op)
        sess.run(simulate_op)
        ts[i+1] = ts[i] + 0.01

plt.plot(ts,xs)
plt.show()
```
